```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
height = 360
width = 640
DEPTH_DTYPE = np.uint16 
```

```python
# Load 1D raw data
depth_raw = np.fromfile('depth.raw', dtype=DEPTH_DTYPE)
#Shape to image
depth_u16 = depth_raw.reshape(height, width)
# Normalize from u16 to u8 to show image
depth_u8 = cv2.normalize(depth_u16, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
plt.imshow(cv2.cvtColor(depth_u8, cv2.COLOR_GRAY2BGR))
```

```python
#define danger zone of 2m (2000mm)
danger_threshhold = 2000
# mask all but danger zone
for i in range(height):
    for j in range(width):
        if (depth_u16[i][j] > danger_threshhold) | (depth_u16[i][j] == 0):
            depth_u16[i][j] = 0
depth_img = cv2.normalize(depth_u16, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
plt.imshow(cv2.cvtColor(depth_img, cv2.COLOR_GRAY2BGR))
```
