```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Here's my mentality for this thread:
# The ground will have a (more-or-less) uniform gradient. Based on the camera angle, we can poll/calculate this.
# We can map this expected distance (depth data) to the pixel height with a linear equation
# If the pixel value is within +- the expected range, it's the ground
# If it's outside that range (by, say 15cm) we can assum it's an object.
# Filter out object pixels and we have values for our cost map
```

```python
# Load Image
image_c = cv2.imread('./data/obstacle.png')
plt.imshow(cv2.cvtColor(image_c, cv2.COLOR_RGB2BGR))
```

```python
#create psuedo depth data for floor
depth_data = cv2.cvtColor(image_c, cv2.COLOR_RGB2GRAY)
height, width = depth_data.shape
for i in range(width):
    for j in range(height):
        depth_data[j][i] = 130 - (j/height)*130 + 20
plt.imshow(cv2.cvtColor(depth_data, cv2.COLOR_RGB2BGR))
```

```python
barrel1 = np.array([ [336,137],
                     [408,167],
                     [462,14],
                     [372,6] ], dtype=np.int32)
barrel2 = np.array([ [61,119],
                     [128,109],
                     [97,0],
                     [19,0] ], dtype=np.int32)
cv2.fillPoly(depth_data, [barrel1], 90)
cv2.fillPoly(depth_data, [barrel2], 103)
cv2.imwrite('./data/pseudo_depth_obstacle.png', depth_data)
plt.imshow(cv2.cvtColor(depth_data, cv2.COLOR_RGB2BGR))
```

```python
for i in range(width):
    for j in range(height):
        a = depth_data[j][i]
        y_val = 130 - (j / height) * 130 + 20
        if (int(a) > y_val - 10) and (int(a) < y_val + 10):
            depth_data[j][i] = 0
depth_data[depth_data > 0] = 255
plt.imshow(cv2.cvtColor(depth_data, cv2.COLOR_RGB2BGR))
```

```python
cv2.imwrite('./processed/depth_costmap.png', depth_data)
```
